### *4.1 QUESTION 1*

Identifying the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

```python
#selecting the city of Paris
paris_df=ecars_df.loc[ecars_df['city']=='Paris']
paris_df
```

```python
#sorting the data by station id, day, hour and  minute
df=paris_df.sort_values(['station_id','day','hour','minute'], ascending=True)
df.head(10)
```

```python
#creating a new column of 'usage' by finding the difference of bluecar successive rows 
#this will be done across every station
df['usage'] = df.groupby('station_id')['bluecar'].diff()
df.tail(20)
```

+ve number means that a car was returned -ve number means that a car was picked 0 means that nothing happened, i.e no car was picked nor returned

```python
#grouping by hour and finding the sum
df.groupby('hour')['usage'].agg(sum)
```

the hour with the highest negative value is the hour most cars were picked which is 1800hrs basically this is 6.00pm 29 cars were picked across the stations
